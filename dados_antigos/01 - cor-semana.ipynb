```python
import numpy as np #lib para cálculos e funções em arrays multidimensionais
import pandas as pd #pandas - libpra manipulação e análise de dados de alta performance
```

## 1 - INGESTÃO: Carregando primeiro Dataset ##   

```python
cor_se_temp = pd.read_csv('cor-semana-covid.csv', sep=';', encoding = "ISO-8859-1",skiprows=[0,1,2,3,12,13,14,15])

print('\nTabela Original - Óbitos por covid / Raça / Semana Epidemiológica')

display(cor_se_temp)
```

## 2 - Tratamento dos dados

```python
## Converter os valores todos para numérico, se der erro, setar como NaN e depois converter pro número 0
cor_se = cor_se_temp.apply(pd.to_numeric, errors='coerce').fillna(0)

print('\nResultado do Primeiro Tratamento - Óbitos por covid / Raça / Semana Epidemiológica')


display(cor_se)
```

```python
## Como eu perdi os dados da coluna 0 (que foram transformados no valor 0), preciso resgatar do Dataset original e sobrescrever.
cor_se['Cor'] = cor_se_temp['Cor']

print('\nResultado do Segundo Tratamento - Óbitos por covid / Raça / Semana Epidemiológica')


display(cor_se)
```

```python
#removendo a primeira e as 2 últimas colunas
cor_se_norm = cor_se.loc[:, cor_se.columns[1:-2]]

#realizando a soma cumulativa através das colunas (da esquerda pra direita ------->)
cor_se_norm = cor_se_norm.cumsum(axis=1)

display(cor_se_norm)


```

```python
#adicionando novamente a coluna das cores
cor_se_norm['Cor'] = cor_se['Cor']
display(cor_se_norm)
```

## 3 - Exportanto resultado

```python
cor_se_norm.to_csv('cor-semana-covid-tratado.csv')
```
